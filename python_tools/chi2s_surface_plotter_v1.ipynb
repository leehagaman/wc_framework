```python
import numpy as np
import matplotlib.pyplot as plt
```

# Chi2 Maps

```python
with open("../TLee_Np0p/constrained_glee_only_data_chi2_map.txt") as f:
    lines = f.readlines()
#with open("../TLee_Np0p/glee_only_data_chi2_map.txt") as f:
#    lines = f.readlines()
chi2_map = np.zeros((101, 101))
for line in lines:
    i = int(line.split(",")[0][1:].strip())
    j = int(line.split(":")[0].split(",")[1][:-2].strip())
    chi2 = float(line.split(":")[1].strip())
    chi2_map[i, j] = chi2
chi2_map = chi2_map.T

plt.figure(figsize=(7, 7))
plt.imshow(chi2_map, origin='lower')
plt.colorbar()
#plt.scatter(15, 5, c="r", label="(15, 5)", marker="*", s=200, edgecolors="k")
plt.scatter(1, 1, c="b", label="(1, 1)", marker="*", s=200, edgecolors="k")
plt.scatter(1e-06, 8.15884, c="r", label="Best-Fit: (1e-06, 8.15884)", marker="*", s=200, edgecolors="k")
plt.legend()
plt.xlabel(r"$x_{\Delta N p}$")
plt.ylabel(r"$x_{\Delta 0 p}$")
plt.title(r"gLEE-only Real Data $\chi^2$ Map")
plt.show()

plt.figure(figsize=(7, 7))
plt.imshow(chi2_map, origin='lower', aspect='auto')
plt.colorbar()
#plt.scatter(15, 5, c="r", label="(15, 5)", marker="*", s=200, edgecolors="k")
plt.scatter(1, 1, c="b", label="(1, 1)", marker="*", s=200, edgecolors="k")
plt.scatter(1e-06, 8.15884, c="r", label="Best-Fit: (1e-06, 8.15884)", marker="*", s=200, edgecolors="k")
plt.legend()
plt.xlabel(r"$x_{\Delta N p}$")
plt.ylabel(r"$x_{\Delta 0 p}$")
plt.xlim(0, 10)
plt.title(r"gLEE-only Real Data $\chi^2$ Map")
plt.show()

print(np.min(chi2_map))
# print i,j of min chi2
print(np.where(chi2_map == np.min(chi2_map)))
```
