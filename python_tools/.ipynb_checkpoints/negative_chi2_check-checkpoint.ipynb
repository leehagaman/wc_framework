```python
import numpy as np
import uproot as uproot
import pandas as pd

import matplotlib.pyplot as plt

from scipy import linalg
```

```python
# getting CVs (used for Asimov and LEE preds)

f = uproot.open(f"../TLee_Np0p/data_inputs/merge.root")

all_mc_preds = []
is_Np_sig = []
is_0p_sig = []

for i in range(1, 25): # 1-18 are uncollapsed mc predictions, 19-24 are ext predictions
    current_preds = list(f[f"histo_{i}"].to_numpy(flow=True)[0][1:])
    all_mc_preds += current_preds # get the overflow bin only
    
    for _ in range(len(current_preds)):
        if i > 18 or i % 3 == 1:
            is_Np_sig.append(0)
            is_0p_sig.append(0)
        elif i % 3 == 2:
            is_Np_sig.append(1)
            is_0p_sig.append(0)
        else:
            is_Np_sig.append(0)
            is_0p_sig.append(1)

all_mc_preds = np.array(all_mc_preds)
is_Np_sig = np.array(is_Np_sig)
is_0p_sig = np.array(is_0p_sig)

print(len(all_mc_preds))
```

```python
collapsing_matrix = np.zeros((68, 272))

# 1gNp bin
collapsing_matrix[0, 0]   = 1. # MC bkg
collapsing_matrix[0, 2]   = 1. # true Np signal
collapsing_matrix[0, 4]   = 1. # true 0p signal
collapsing_matrix[0, 204] = 1. # ext bkg
# 1gNp overflow
collapsing_matrix[1, 1]   = 1.
collapsing_matrix[1, 3]   = 1.
collapsing_matrix[1, 5]   = 1.
collapsing_matrix[1, 205] = 1.
# 1g0p bin
collapsing_matrix[2, 6]   = 1.
collapsing_matrix[2, 8]   = 1.
collapsing_matrix[2, 10]  = 1.
collapsing_matrix[2, 206] = 1.
# 1g0p overflow
collapsing_matrix[3, 7]   = 1.
collapsing_matrix[3, 9]   = 1.
collapsing_matrix[3, 11]  = 1.
collapsing_matrix[3, 207] = 1.

# NC Pi0 Np
for i in range(16):
    collapsing_matrix[4 + i, 12 + i] = 1.
    collapsing_matrix[4 + i, 12 + 16 + i] = 1.
    collapsing_matrix[4 + i, 12 + 2*16 + i] = 1.
    collapsing_matrix[4 + i, 208 + i] = 1.

# NC Pi0 0p
for i in range(16):
    collapsing_matrix[4+16 + i, 12+3*16 + i] = 1.
    collapsing_matrix[4+16 + i, 12+4*16 + i] = 1.
    collapsing_matrix[4+16 + i, 12+5*16 + i] = 1.
    collapsing_matrix[4+16 + i, 208+16 + i] = 1.

# numuCC Np
for i in range(16):
    collapsing_matrix[4+2*16 + i, 12+6*16 + i] = 1.
    collapsing_matrix[4+2*16 + i, 12+7*16 + i] = 1.
    collapsing_matrix[4+2*16 + i, 12+8*16 + i] = 1.
    collapsing_matrix[4+2*16 + i, 208+2*16 + i] = 1.
    
# numuCC 0p
for i in range(16):
    collapsing_matrix[4+3*16 + i, 12+9*16 + i] = 1.
    collapsing_matrix[4+3*16 + i, 12+10*16 + i] = 1.
    collapsing_matrix[4+3*16 + i, 12+11*16 + i] = 1.
    collapsing_matrix[4+3*16 + i, 208+3*16 + i] = 1.  
    
#collapsing_matrix = np.transpose(collapsing_matrix)

collapsing_matrix.shape
```

```python
asimov_data_uncollapsed = all_mc_preds
lee_scalings = [4.0803701, 8946089.3] # found from best fit point that got a negative chi^2 with detvar turned off
lee_uncollapsed = all_mc_preds + (lee_scalings[0] - 1.) * is_Np_sig * all_mc_preds + (lee_scalings[1] - 1.) * is_0p_sig * all_mc_preds

plt.figure(figsize=(10,7))
plt.plot(asimov_data_uncollapsed, label="CV")
plt.plot(lee_uncollapsed, label="lee scale point")
plt.xlabel("uncollapsed bin index")
plt.yscale("log")
plt.ylim(0.001, 1e10)
plt.legend()
plt.show()

asimov_data_collapsed = collapsing_matrix.dot(asimov_data_uncollapsed)
lee_collapsed = collapsing_matrix.dot(lee_uncollapsed)

plt.figure(figsize=(10,7))
plt.plot(asimov_data_collapsed, label="CV")
plt.plot(lee_collapsed, label="lee scale point")
plt.xlabel("collapsed bin index")
plt.yscale("log")
plt.ylim(0.001, 1e10)
plt.legend()
plt.show()
```
