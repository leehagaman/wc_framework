```python
with open("../TLee_Np0p/cv_and_cov_csv_outputs/total_cov_no_data_stat_diag.csv") as f:

    lines = f.readlines()

    dim = len(lines[0].split(",")[:-1])

    temp = np.zeros([dim, dim])
    for i in range(dim):
        temp[i, :] = list(map(float, lines[i].split(",")[:-1]))

plt.figure()
plt.imshow(temp, origin="lower")
plt.colorbar()
plt.title("Total Cov, no data stat diag")
plt.show()

plt.figure()
plt.imshow(temp, origin="lower", norm=mpl.colors.LogNorm())
plt.colorbar()
plt.title("Total Cov, no data stat diag")
plt.show()

linthresh = np.min(np.abs(temp[np.nonzero(temp)]))
plt.figure()
plt.imshow(temp, origin="lower", norm=mpl.colors.SymLogNorm(linthresh=linthresh))
plt.colorbar()
plt.title("Total Cov, no data stat diag")
plt.show()
```

```python
filenames = [

    "total_cov_no_data_stat_diag",

    "Asimov_data_stat_diag_cov",
    "real_data_stat_cor_cov",
    "mc_stat_diag_cov",
    "mc_stat_cor_cov",

    "flux_cov",
    "Xs_cov",
    "detector_cov",
    "additional_cov",

    # not used, just checking
    "reweight_cor_cov",
    "reweight_cov",
    "mc_stat_cov",
]

# these all get replaced by the globals statement in the loop
total_cov_no_data_stat_diag = -1
Asimov_data_stat_diag_cov = -1
real_data_stat_cor_cov = -1
mc_stat_diag_cov = -1
mc_stat_cor_cov = -1
flux_cov = -1
Xs_cov = -1
detector_cov = -1
additional_cov = -1
reweight_cor_cov = -1
reweight_cov = -1
mc_stat_cov = -1

names = [
    "Total Cov, no data stat diag",

    "Asimov Data Stat Diag Cov",
    "Real Data Stat Cor Cov",
    "MC Stat Diag Cov",
    "MC Stat Cor Cov",

    "Flux Cov",
    "Xs Cov",
    "Detector Cov",
    "Additional Cov",

    "(Unused) Reweight Cor Cov",
    "(Unused) Reweight Cov",
    "(Unused) MC Stat Cov",
]


for t in range(len(filenames)):

    filename = filenames[t]
    name = names[t]

    with open(f"../TLee_Np0p/cv_and_cov_csv_outputs/{filename}.csv") as f:
        lines = f.readlines()
        dim = len(lines[0].split(",")[:-1])
        curr_cov = np.zeros([dim, dim])
        for i in range(dim):
            curr_cov[i, :] = list(map(float, lines[i].split(",")[:-1]))

        globals()[filename] = curr_cov

    plt.figure()
    plt.imshow(curr_cov, origin="lower")
    plt.colorbar()
    plt.title(name)
    plt.show()
```
