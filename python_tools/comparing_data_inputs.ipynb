```python
for i in range(0, 100):
    print(f"new mc_stat file, {i}")
    with open(old_data_inputs_dir + f"/mc_stat/{i}.log", "r") as f:
        old_lines = f.readlines()
    with open(new_data_inputs_dir + f"/mc_stat/{i}.log", "r") as f:
        new_lines = f.readlines()

    all_old_nums = []
    for old_line in old_lines:
        old_nums = [float(x) for x in old_line.split()]
        all_old_nums.append(old_nums)
    all_new_nums = []
    for new_line in new_lines:
        new_nums = [float(x) for x in new_line.split()]
        all_new_nums.append(new_nums)

    for i in range(len(all_old_nums)):
        #print("    old:", all_old_nums[i])
        #print("    new:", all_new_nums[i])
        print("    new-old:", np.array(all_new_nums[i]) - np.array(all_old_nums[i]))
```

```python
mpl.rcParams['font.size'] = 14

labels = [
    "LYDown", # 1
    "LYRayleigh", # 2
    "Recomb2", # 3
    "SCE", # 4
    "dEdx", # 5 # NOT USED!
    "WMThetaXZ", # 6
    "WMThetaYZ", # 7
    "WMX", # 8
    "WMYZ", # 9
    "LYatt", # 10
]

def my_step(values, label, c=None, ls="-"):
    values = list(values)
    values = values + [values[-1]]
    if c is not None:
        plt.step(np.arange(len(values)), values, where="post", label=label, lw=2, c=c, ls=ls)
    else:
        plt.step(np.arange(len(values)), values, where="post", label=label, lw=2, ls=ls)
```

```python
old_diag_errs = {}

for i in range(1, 11):
    label = labels[i-1]
    if i == 5:
        continue
    frac_cov = uproot.open(f"{old_data_inputs_dir}/hist_rootfiles/DetVar/cov_{label}.root")[f"frac_cov_det_mat_{i}"].member("fElements").reshape((288, 288))
    #plt.figure()
    #plt.imshow(frac_cov, origin="lower", aspect="auto", norm=mpl.colors.LogNorm())
    #plt.colorbar()
    #plt.show()
    old_diag_errs[label] = np.sqrt(np.diag(frac_cov))

plt.figure(figsize=(10, 6))
for i in range(1, 11):
    if i == 5:
        continue
    label = labels[i-1]
    my_step(old_diag_errs[label], label)
plt.legend()
plt.title("old")
plt.show()

old_even_bin_diag_errs = {}
for k, v in old_diag_errs.items():
    old_even_bin_diag_errs[k] = v[::2]
```

```python
new_diag_errs = {}

for i in range(1, 11):
    label = labels[i-1]
    if i == 5:
        continue
    frac_cov = uproot.open(f"{new_data_inputs_dir}/hist_rootfiles/DetVar/cov_{label}.root")[f"frac_cov_det_mat_{i}"].member("fElements").reshape((288, 288))
    #plt.figure()
    #plt.imshow(frac_cov, origin="lower", aspect="auto", norm=mpl.colors.LogNorm())
    #plt.colorbar()
    #plt.show()
    new_diag_errs[label] = np.sqrt(np.diag(frac_cov))

plt.figure(figsize=(10, 6))
for i in range(1, 11):
    if i == 5:
        continue
    label = labels[i-1]
    my_step(new_diag_errs[label], label)
plt.legend()
plt.title("new")
plt.show()

new_even_bin_diag_errs = {}
for k, v in new_diag_errs.items():
    new_even_bin_diag_errs[k] = v[::2]
```

```python
plt.figure(figsize=(10, 6))
for i in range(1, 11):
    if i == 5:
        continue
    label = labels[i-1]
    my_step(new_even_bin_diag_errs[label], label)
plt.legend()
plt.xlim(0, 12)
plt.ylim(0, 1)
plt.axvline(3, c="k", ls="--")
plt.axvline(6, c="k", ls="--")
plt.axvline(9, c="k", ls="--")
plt.xticks(np.arange(12)+0.5, [
    "WC 1gNp bkg", "WC 1gNp\nNp sig", "WC 1gNp\n0p sig", 
    "WC 1g0p bkg", "WC 1g0p\nNp sig", "WC 1g0p\n0p sig",
    "gLEE 1g1p bkg", "gLEE 1g1p\nNp sig", "gLEE 1g1p\n0p sig",
    "gLEE 1g0p bkg", "gLEE 1g0p\nNp sig", "gLEE 1g0p\n0p sig",
    ], rotation=90, fontsize=10)
plt.ylabel("Fractional Error")
plt.title("New Data Inputs")
plt.show()

plt.figure(figsize=(10, 6))
for i in range(1, 11):
    if i == 5:
        continue
    label = labels[i-1]
    my_step(old_even_bin_diag_errs[label], label)
plt.legend()
plt.xlim(0, 12)
plt.ylim(0, 1)
plt.axvline(3, c="k", ls="--")
plt.axvline(6, c="k", ls="--")
plt.axvline(9, c="k", ls="--")
plt.xticks(np.arange(12)+0.5, [
    "WC 1gNp bkg", "WC 1gNp\nNp sig", "WC 1gNp\n0p sig", 
    "WC 1g0p bkg", "WC 1g0p\nNp sig", "WC 1g0p\n0p sig",
    "gLEE 1g1p bkg", "gLEE 1g1p\nNp sig", "gLEE 1g1p\n0p sig",
    "gLEE 1g0p bkg", "gLEE 1g0p\nNp sig", "gLEE 1g0p\n0p sig",
    ], rotation=90, fontsize=10)
plt.ylabel("Fractional Error")
plt.title("Old Data Inputs")
plt.show()
```
