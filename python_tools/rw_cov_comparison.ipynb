```python
import numpy as np
import uproot as uproot

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
f = uproot.open(f"../TLee_Np0p/data_inputs/merge.root")
uncollapsed_all_mc_preds = []
for i in range(1, 25): # 1-18 are uncollapsed mc predictions, 19-24 are ext predictions
    current_preds = list(f[f"histo_{i}"].to_numpy(flow=True)[0][1:])
    uncollapsed_all_mc_preds += current_preds # get the overflow bin only
uncollapsed_all_mc_preds = np.array(uncollapsed_all_mc_preds)
```

```python
collapsing_matrix = np.zeros((4, 12))

# collapsing 1gNp bin
collapsing_matrix[0,0] = 1. # bkg
collapsing_matrix[0,2] = 1. # Np sig
collapsing_matrix[0,4] = 1. # 0p sig

# collapsing 1gNp overflow bin
collapsing_matrix[1,1] = 1. # bkg
collapsing_matrix[1,3] = 1. # Np sig
collapsing_matrix[1,5] = 1. # 0p sig

# collapsing 1g0p bin
collapsing_matrix[2,6] = 1. # bkg
collapsing_matrix[2,8] = 1. # Np sig
collapsing_matrix[2,10] = 1. # 0p sig

# collapsing 1g0p overflow bin
collapsing_matrix[3,7] = 1.  # bkg
collapsing_matrix[3,9] = 1.  # Np sig
collapsing_matrix[3,11] = 1. # 0p sig

collapsed_all_mc_preds_nc_delta = [
    uncollapsed_all_mc_preds[0] + uncollapsed_all_mc_preds[2] + uncollapsed_all_mc_preds[4], # 1gNp bin (bkg + Np sig + 0p sig)
    uncollapsed_all_mc_preds[1] + uncollapsed_all_mc_preds[3] + uncollapsed_all_mc_preds[5], # 1gNp overflow
    uncollapsed_all_mc_preds[6] + uncollapsed_all_mc_preds[8] + uncollapsed_all_mc_preds[10], # 1g0p bin (bkg + Np sig + 0p sig)
    uncollapsed_all_mc_preds[7] + uncollapsed_all_mc_preds[9] + uncollapsed_all_mc_preds[11], # 1g0p overflow
]
```

```python
diagonal_sigmas_uncollapsed = []
diagonal_sigmas_nc_delta_collapsed = []

full_flux_cov_mat = np.zeros((272, 272))

# flux uncertainties

diagonal_sigmas_uncollapsed.append([])
diagonal_sigmas_nc_delta_collapsed.append([])
for idx in range(1,17):
    f_cov = uproot.open(f"../TLee_Np0p/data_inputs/hist_rootfiles/XsFlux/cov_{idx}.root")

    cov_mat = f_cov[f"cov_xf_mat_{idx}"].member("fElements")
    cov_mat = cov_mat.reshape((272, 272))
    
    full_flux_cov_mat += cov_mat
    
    print(full_flux_cov_mat[50,50])
    
for i in range(272):
    diagonal_sigmas_uncollapsed[-1].append(np.sqrt(full_flux_cov_mat[i,i]))

cov_mat_nc_delta = full_flux_cov_mat[:12,:12]
manually_collapsed_cov_mat_nc_delta = np.matmul(collapsing_matrix, np.matmul(cov_mat_nc_delta, np.transpose(collapsing_matrix)))

for i in range(4):
    diagonal_sigmas_nc_delta_collapsed[-1].append(np.sqrt(manually_collapsed_cov_mat_nc_delta[i,i]))

    


for idx in range(17, 20): # Xs and reweighting uncertainties
    diagonal_sigmas_uncollapsed.append([])
    diagonal_sigmas_nc_delta_collapsed.append([])

    # 17 is Xs
    # 18 is uncorrelated reweighting uncertainty
    # 19 is correlated reweighting uncertainty
    
    f_cov = uproot.open(f"../TLee_Np0p/data_inputs/hist_rootfiles/XsFlux/cov_{idx}.root")
    #print(f_rw_cov.keys())

    cov_mat = f_cov[f"cov_xf_mat_{idx}"].member("fElements")
    cov_mat = cov_mat.reshape((272, 272))
    
    for i in range(272):
        diagonal_sigmas_uncollapsed[-1].append(np.sqrt(cov_mat[i,i]))

    cov_mat_nc_delta = cov_mat[:12,:12]
    manually_collapsed_cov_mat_nc_delta = np.matmul(collapsing_matrix, np.matmul(cov_mat_nc_delta, np.transpose(collapsing_matrix)))

    for i in range(4):
        diagonal_sigmas_nc_delta_collapsed[-1].append(np.sqrt(manually_collapsed_cov_mat_nc_delta[i,i]))

    

```

```python
lw = 4
plt.rcParams.update({'font.size': 12})

plt.figure(figsize=(10,7))
plt.hist([_+0.5 for _ in range(12)], weights=np.nan_to_num(np.array(diagonal_sigmas_uncollapsed[0][:12])/np.array(uncollapsed_all_mc_preds[:12])), 
         bins=12, range=(0, 12), histtype="step", lw=lw, label="Flux Uncertainty", color="r")
plt.hist([_+0.5 for _ in range(12)], weights=np.nan_to_num(np.array(diagonal_sigmas_uncollapsed[1][:12])/np.array(uncollapsed_all_mc_preds[:12])), 
         bins=12, range=(0, 12), histtype="step", lw=lw, label="Xs Uncertainty", color="b")
plt.hist([_+0.5 for _ in range(12)], weights=np.nan_to_num(np.array(diagonal_sigmas_uncollapsed[2][:12])/np.array(uncollapsed_all_mc_preds[:12])), 
         bins=12, range=(0, 12), histtype="step", lw=lw, label="Uncorrelated Reweighting Uncertainty", color="tab:olive")
plt.hist([_+0.5 for _ in range(12)], weights=np.nan_to_num(np.array(diagonal_sigmas_uncollapsed[3][:12])/np.array(uncollapsed_all_mc_preds[:12])), 
         bins=12, range=(0, 12), histtype="step", lw=lw, ls="dotted", label="Correlated Reweighting Uncertainty", color="gold")
plt.xlabel("bin number")
plt.ylabel("Fractional Uncertainty (fraction of events for one sigma deviation)")
plt.legend()
plt.show()

plt.figure(figsize=(10,7))
plt.hist([_+0.5 for _ in range(4)], weights=np.nan_to_num(np.array(diagonal_sigmas_nc_delta_collapsed[0][:12])/np.array(collapsed_all_mc_preds_nc_delta)), 
         bins=12, range=(0, 4), histtype="step", lw=lw, label="Flux Uncertainty", color="r")
plt.hist([_+0.5 for _ in range(4)], weights=np.nan_to_num(np.array(diagonal_sigmas_nc_delta_collapsed[1][:12])/np.array(collapsed_all_mc_preds_nc_delta)), 
         bins=12, range=(0, 4), histtype="step", lw=lw, label="Xs Uncertainty", color="b")
plt.hist([_+0.5 for _ in range(4)], weights=np.nan_to_num(np.array(diagonal_sigmas_nc_delta_collapsed[2][:12])/np.array(collapsed_all_mc_preds_nc_delta)), 
         bins=12, range=(0, 4), histtype="step", lw=lw, label="Uncorrelated Reweighting Uncertainty", color="tab:olive")
plt.hist([_+0.5 for _ in range(4)], weights=np.nan_to_num(np.array(diagonal_sigmas_nc_delta_collapsed[3][:12])/np.array(collapsed_all_mc_preds_nc_delta)), 
         bins=12, range=(0, 4), histtype="step", lw=lw, ls="dotted", label="Correlated Reweighting Uncertainty", color="gold")
plt.xlabel("bin number")
plt.ylabel("Fractional Uncertainty (fraction of events for one sigma deviation)")
plt.legend()
plt.show()


```
