```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
arr = np.loadtxt("2d_ratio_0p1pi0.csv", delimiter=",", skiprows=1)

plt.rcParams.update({'font.size': 16})
plt.figure(figsize=(10, 6))
plt.scatter(arr[:, 0], arr[:, 1], c=arr[:, 2], s=5, norm=mpl.colors.Normalize(vmin=0, vmax=2), cmap="RdBu")
plt.colorbar()
plt.xlabel("pi0 energy")
plt.ylabel("pi0 costheta")
plt.title("0p 1pi0")
plt.show()

arr = np.loadtxt("2d_ratio_Np1pi0.csv", delimiter=",", skiprows=1)

plt.rcParams.update({'font.size': 16})
plt.figure(figsize=(10, 6))
plt.scatter(arr[:, 0], arr[:, 1], c=arr[:, 2], s=5, norm=mpl.colors.Normalize(vmin=0, vmax=2), cmap="RdBu")
plt.colorbar()
plt.xlabel("pi0 energy")
plt.ylabel("pi0 costheta")
plt.title("Np 1pi0")
plt.show()

```
